```python
# Tukey 함수 구현

import numpy as np

def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr*1.5)
    upper_bound = q3 + (iqr*1.5)
    return np.where((data > upper_bound)|(data < lower_bound))

outliers = outliers_iqr(bike_data.Distance)
```

### 가설 검정과 t-검정

 : 같은 모집단에서 나온 다른 샘플의 **평균**이 통계적으로 의미있게 비슷한 값인지 판단

```python
bike_data2.Gu.value_counts()
```

```python
y_gu = bike_data2[bike_data2.Gu == '영등포구']
m_gu = bike_data2[bike_data2.Gu == '마포구']
```

#### 1. Levene 의 등분산 검정 : 등분산인지 확인

```python
from scipy import stats

stats.levene(y_gu.Distance, m_gu.Distance)
```

- 검정통계량 statistics = 3.56
- pvalue = 0.059

P-value > 0.05 이면 등분산, P-value < 0.05 이면 등분산이 아닌 것으로 판단

```python
np.mean(y_gu.Distance)
```

```python
np.mean(m_gu.Distance)
```

#### 2. 가설검정

- 가설검정 
    - 귀무가설 정의 : '마포구와 영등포구의 공유자전거 이동거리의 평균은 같다'
    - 대립가설 정의 : 귀무 가설에 대한 반대 가설
    - 표본통계량 계산
    - P-value와 유의수준과 비교 : P-value가 유의수준보다 작으면 귀무가설 기각

#### 3. t-검정 : 두 집단 간의 평균에 대한 검정을 통계적 유의성으로 검정하는 방법

- Two Sample t-test (이표본 t검정) : 표본의 평균이 같은지 검정
    - 두개의 데이터 세트를 비교
    - ex) 두 반의 영어 성적 비교
    

- One Sample t-test (일표본 t검정) : 표본의 평균이 모집단 평균과 같은지 검정
    - 한쪽은 알려진 값으로 평균값이 제시, 다른 한쪽은 데이터로 제시되어 평균을 구함.
    - 알려진 평균값과 계산한 표본 평균을 비교함
    - ex) LED 조명의 평균 수명시간
    
    
- Paired t-test (대응표본 t검정) : 대응하는 두 표본의 평균 차이가 특정값과 같은지 검정
    - 두 데이터 세트가 주어지되, Before - After 같이 짝이 지어짐.
    - ex) 영어성적 10점 향상 공부법

예시> Two sample t-test

```python
# equal_var=True : 등분산으로 판단 (by.Levene 등분산 검정)
stats.ttest_ind(y_gu.Distance, m_gu.Distance, equal_var = True)
```

'e-05' : 0.05보다 꽤 많이 작은 값의 수치로 표시

-> P-value가 0.05보다 작다는 것으로 이해

-> 귀무가설 기각 : 두 구의 이동거리 평균은 같지 않다

- P-value
    - 감당할 수 있는 리스크에 대한 확률
    - 귀무가설이 참이라는 전제 하에 검정통계량보다 큰 값이 나타날 확률


- 유의수준과 P-value 간의 비교

    - 유의수준 0.05 : 통계적으로 의미있는 차이의 기준값
    
    - P-value : 귀무가설이 참이라는 가정 하에 현재 두 평균값 사이의 차이가 나타날 확률
    
    
- P-value가 유의수준보다 작은 값이라면 너무 낮은 확률이므로 귀무가설 기각.


- P-value는 0.05로 고정된 기준 값이 아닌 풀고자 하는 문제에 따라 스스로 정하는 값.

    - 생명을 다루는 문제 : 0.01, 0.001도 가능

### 분산분석 (ANOVA)

: 2개 이상의 데이터 그룹을 비교하기 위해 사용

- 두 개의 평균이 통계적으로 다르다면 두 평균의 분산도 겹치지 않는다

#### 분산분석의 단계
1. 등분산분석 : 여러 그룹의 분산이 같은가를 확인
    - 가설 검정 방법대로 귀무 가설 정의 '모든 그룹의 분산은 같다'
    - 유의수준(0.05)과 p-value를 비교 : p-value가 유의수준보다 작으면 귀무가설 기각
    - 등분산임을 가정할 수 있으면 분산분석 가능. 그렇지 않으면 다른 방법 수행.


2. 분산분석(One Way ANOVA)
    - 귀무가설의 정의 '모든 그룹의 평균은 같다'
    - 대립가설 정의
    - 유의수준(0.05)과 p-value를 비교 : p-value가 유의수준보다 작으면 귀무가설 기각 
    - 기각 = 어떤 그룹의 평균은 같지 않다.
    

3. 사후분석
    - 그룹별 평균을 비교해 평균이 같은 그룹과 다른 그룹을 구분하여 알려줌.

- Bartlett's Test 등분산 분석

```python
s_gu = bike_data2[bike_data2.Gu == '서초구']
d_gu = bike_data2[bike_data2.Gu == '동대문구']
e_gu = bike_data2[bike_data2.Gu == '은평구']
```

```python
from scipy import stats

stats.bartlett(y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance)

# p-value < 0.05 - 귀무가설 기각 : 모든 그룹이 등분산은 아니다.
```

- 등분산이 나오지 않는다면?

1. **비모수적** 분석 방법 사용
    - 평균, 분산, 표준편차 등 통계량에 근거한 추론 방법을 사용하지 않는 것.
    
    
2. 데이터에 보완할 점 검토

2. 분산분석 : 등분산 시에만 실행해야 하지만 등분산을 가정하고 실습 진행..

```python
stats.f_oneway(y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance)

# p-value < 0.05 - 귀무가설 기각 : 모든 그룹의 평균이 같은 것은 아니다.
```

```python
plot_data =[y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance]

# showfliers=False : 이상치 제외
plt.boxplot(plot_data, showfliers=False)
plt.show()
```

3. 사후분석 : Tukey HSD (평균이 같은 그룹과 다른 그룹을 나누어 볼 수 있는 방법 중 하나)

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(bike_data2.Distance, bike_data2.Gu)
hsd.summary()
```

- p-adj < 0.05 : reject 값이 True
- p-adj > 0.05 : reject 값이 False >> '평균이 같다'라는 귀무가설을 Reject하지 못한다.

### 카이제곱 검정(Chi-Square Test)

: 정규분포의 분산에 대한 확률분포

: 데이터가 평균과 얼마나 떨어져있는지 비교 가능

: 하나의 범주형 데이터가 다른 범주형 데이터에 영향을 미치는지를 확인

: 비교하려는 대상이 수치형 대상이 아닌 경우 >> 비율로 비교, 판단



- 독립성 검정


ex) 성별은 안드로이드폰/아이폰 사용에 영향이 있을까?, 연령은 멤버십 가입 형태에 영향이 있을까?


- 기타 : 적합도 검정, 동질성 검정 등

- 가설검정 방법 활용
    - '기대하는 값이 같다'는 귀무가설 설정
    - 조사 결과 5% 확률 미만의 값 >> 평균으로부터 너무 떨어져 있다.
    - 확률적으로 유의한 범위에서 차이가 크다

```python
from scipy.stats import chi2_contingency

crosstab = pd.crosstab(bike_data2.Age_Group, bike_data2.Membership_type)
chi2_contingency(crosstab)
```

```python
result = chi2_contingency(crosstab)
print('Chi2 Statistic : {}, p-value : {}'.format(result[0], result[1]))
```

- 귀무가설 : Age_Group과 Membership_type은 독립적이다.
- p-value < 0.05 : 가설 기각 = Age_Group과 Membership_type은 독립적이지 않다 = 관련이 있다!

#### 회귀분석 함수 1 : scipy패키지의 stats 모듈 내 함수

```python
stats.linregress(bike_weather.temp, bike_weather.Count)
```

- 선형회귀식을 구성하는 값
    - slope=38.793750423978715
    - intercept=-320.14608008876144
    
    
- 선형회귀식
    - Count = 38.79*temp - 320.146


- R 제곱으로 활용(결정계수) : 회귀식을 판단하는 중요한 결과값
    - rvalue=0.4384055405346409
    
    
- 회귀분석 귀무가설 '회귀관계가 없다/회귀식의 기울기가 0이다'
    - pvalue=0.00016471489031703914 >> 귀무가설 기각 = 회귀관계는 유의미하게 존재
    

- 표준 오차
    - stderr=9.71628875360514

```python
# R제곱(결정계수) : 회귀식이 실제 관찰된 값을 얼마나 설명하는지를 의미
slope, intercept, r_value, p_value, stderr = stats.linregress(bike_weather.temp, bike_weather.Count)
print("R-squared : %f" %r_value**2)
```

#### R제곱(결정계수) : 설명력

- 0에서 1사이 값으로 나타남, 1에 가까울 수록 설명력이 높아 좋은 회귀식임을 의미


- 회귀식이 실제 관찰된 값을 얼마나 설명하는지를 의미


- 데이터 세트에 있는 x값을 회귀식에 넣었을 때 계산된 y값이 실제 데이터에 존재하는 y값 대비 얼마나 되는지를 말함

#### 회귀분석 함수 2 : statsmodels 패키지의 회귀분석 함수

```python
import statsmodels.api as sm

X0 = bike_weather.temp
X1 = sm.add_constant(X0)
y = bike_weather.Count
model = sm.OLS(y, X1)
result = model.fit()
print(result.summary())
```

- R-squared: 0.192
    - R제곱값


- Prob (F-statistic):           0.000165
    - F검정 통계량의 추정치 
    - 회귀식 전체에 대한 통계적 유의성을 검정한 p-value
    
    
- const : 절편 / temp : 독립변수로 사용된 x값
- 두 값의 coef : 회귀식의 절편과 기울기
    - 회귀식 : count = 38.7938*temp - 320.1461
    
 
- x변수들의 p-value에 해당하는 값 (현재는 x가 temp 하나뿐임)
    - x 독립변수가 통계적으로 유의하게 종속변수 y의 변동에 영향을 주는지 확인
    - P > |t| : 0.000 = 독립변수 temp가 종속변수 count의 변화에 기울기 coef만큼 영향을 준다는 것이 '통계적으로 유의하게 판단됨'
